```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate
```

```python
# 문서 로드
loader = PyMuPDFLoader("/content/drive/MyDrive/1706.03762v7.pdf")
docs = loader.load()
```

```python
docs[14]
```

```python
print(docs[1].page_content)
```

```python
# 문서 분할
# print(type(docs[0]))

# docs = list(map(str, docs))
# list(map((lambda x: len(str(x), docs))))
# [len(document) for document in docs]
```

```python
# 문서 분할
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
split_documents = text_splitter.split_documents(docs)
```

```python
type(split_documents)
```

```python
len(split_documents)
```

```python
print(list(map((lambda x: len(str(x))), split_documents)))
```

```python
 # 벡터DB 구축
 vectorstore = FAISS.from_documents(split_documents, embedding=embeddings)
```

```python
# 검색기 생성
retriever = vectorstore.as_retriever()
```

```python
# 프롬프트 생성
prompt = PromptTemplate.from_template(
    """너는 딥러닝에 대해서 지명한 박사야. 논문을 입력해서 전달하면 그에 해당하는 질문에 답변해줘.
    답변은 대학생에게 설명하듯이 해주고 한국어로 해줘.
    최대한 너 스스로 생각하기보다 문서 내에서 답변을 찾아줘.

#Question:
{question}
#Context:
{context}

#Answer:"""
)
```

```python
print(prompt.template)
```

```python
# 언어모델(sLLM) 가져오기
```

```python
llm_chat = ChatUpstage(api_key=api_key)
```

```python
chain = (
    {'context': retriever, 'question': RunnablePassthrough()}
    | prompt
    | llm_chat
    # | StrOutputParser()
)
```

```python
question = input()
response = chain.invoke(question)
print(response)
```
